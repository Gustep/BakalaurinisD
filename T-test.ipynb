```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import ttest_ind
import seaborn as sns
```

```python
table_path = "visi_iverciai.csv"
aktyvumo_stulpelis = 'Aktyvumas'
iverciu_stulpelis = 'DLKcat_mut'

table = pd.read_csv(table_path)
aktyvios = table[table[aktyvumo_stulpelis]==1]
neaktyvios = table[table[aktyvumo_stulpelis]==0]

ttest_ind(aktyvios[iverciu_stulpelis], neaktyvios[iverciu_stulpelis], equal_var=False)
```

# Plots

```python
table_path = "visi_iverciai.csv"                        # failo pavadinimas
aktyvumo_stulpelis = 'Aktyvumas'                         # stulpelio pavadinimas
iverciai = ['CatPred']               # įverčių stulpelių pavadinimai!

# transformuojami duomenys kad būtų patogiau grafiką braižyti
table = pd.read_csv(table_path)
table_norm = table.copy()[iverciai + [aktyvumo_stulpelis]]
table_norm[iverciai] = (table[iverciai] - table[iverciai].mean()) / table[iverciai].std()
long_table = table_norm.melt(id_vars=aktyvumo_stulpelis)


plt.figure(figsize=(10, 5))
sns.violinplot (data=long_table,
            x='variable',
            y='value',
            hue=aktyvumo_stulpelis,
            palette='Spectral'
)


plt.title('Aktyvių ir neaktyvių sekų įverčių pasiskirstymas')
plt.xlabel('Įrankis')
plt.ylabel('Įverčių dydis (standartizuotas)')
plt.show()
```

```python
table_path = "visi_iverciai.csv"                        # failo pavadinimas
aktyvumo_stulpelis = 'Aktyvumas'                         # stulpelio pavadinimas
iverciai = ['DLKcat_nat', 'DLKcat',"CatPred"]               # įverčių stulpelių pavadinimai!

# transformuojami duomenys kad būtų patogiau grafiką braižyti
table = pd.read_csv(table_path)
table_norm = table.copy()[iverciai + [aktyvumo_stulpelis]]
table_norm[iverciai] = (table[iverciai] - table[iverciai].mean()) / table[iverciai].std()
long_table = table_norm.melt(id_vars=aktyvumo_stulpelis)


plt.figure(figsize=(10, 5))
sns.boxplot (data=long_table,
            x='variable',
            y='value',
            hue=aktyvumo_stulpelis,
            palette='Spectral'
)


plt.title('Aktyvių ir neaktyvių sekų įverčių pasiskirstymas')
plt.xlabel('Įrankis')
plt.ylabel('Įverčių dydis (standartizuotas)')
plt.show()
```
