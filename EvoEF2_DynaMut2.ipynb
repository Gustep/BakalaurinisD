```python
# naudojimas
def run_evoef2(filepath: str, verbose: bool = True) -> float:
  assert filepath.endswith('.pdb'), "Please provide PDB file"
  filestem = Path(filepath).stem
  if verbose:
    print(f'Processing {filestem} structure...')
  result_filepath = f'evoef_result_{filestem}.txt'
  command = f'./EvoEF2dir/EvoEF2 --command=ComputeStability --pdb="{filepath}" > {result_filepath}'
  result = subprocess.run(command, shell=True, text=True, capture_output=True)
  if result.returncode != 0:
    print(f'An error happened while running EvoEF2. Please make sure the file exists.\nThe error: {result.stderr}\nAlso check the generated file.')
    return
  with open(result_filepath, 'r') as f:
    total_energy = float(f.read().splitlines()[-3].split(' ')[-1].strip())
  if verbose:
    print(f'Calculated total energy: {total_energy}')
  return total_energy
```

# Paleisti visiems PDB failams nurodytame aplanke ir gauti vieną CSV failą su visais skaičiavimais

```python
pdb_zip_file = "Mut_PDB.zip"


# išarhyvuoti failus į aplanką
with zipfile.ZipFile(pdb_zip_file, 'r') as zip_ref:
    zip_ref.extractall()

# paruošti zodyna saugoti energijas
energies = {}

mut_pdb_folder_path = Path(pdb_zip_file.split('.zip')[0])
for subfolder in mut_pdb_folder_path.glob("*"):
  energies[subfolder.name] = {}
  for pdb_file in tqdm(subfolder.glob('*.pdb')):
    calculated_energy = run_evoef2(filepath=str(pdb_file), verbose=True)
    energies[subfolder.name][pdb_file.stem] = calculated_energy
```

# DynaMut2 automatinė versija

```python
import os, json
import zipfile
from tqdm import tqdm
from time import sleep

zip_with_structures = 'pdbs_cif.zip'
zip_with_mutations = 'mut_lists.zip'
```

```python
dynamut2_link = "https://biosig.lab.uq.edu.au/dynamut2/api/prediction_mm"
curl_template = "curl {dynamut2_link} -X POST -i -F pdb_file=@{pdb_file} -F mutations_list=@{mut_list}"

curl_job_ids = {}

# išarhyvuoti failus į aplanką
with zipfile.ZipFile(zip_with_structures, 'r') as zip_ref:
    zip_ref.extractall()
with zipfile.ZipFile(zip_with_mutations, 'r') as zip_ref:
    zip_ref.extractall()

mut_list_dir = zip_with_mutations.rsplit('.zip', 1)[0]
pdb_dir = zip_with_structures.rsplit('.zip', 1)[0]
mut_lists = os.listdir(mut_list_dir)

print(f'Bus {len(mut_lists)} užklausų\n')

# paleidžiama kiekvienai struktūrai su pauzėm tarp užklausų
for m in mut_lists:
  seq_name = m.split('_')[0]
  mut_list = os.path.join(mut_list_dir, m)
  pdb_file = [x for x in os.listdir(pdb_dir) if (seq_name in x) and x.endswith('.pdb')]
  if len(pdb_file) == 0:
    print(f'Nėra struktūros {seq_name} sekai?')
    continue
  elif len(pdb_file) > 1:
    print(f'Rastos kelios {seq_name} struktūros: {pdb_file}\nBus naudojama pirma iš rastų.')
  pdb_file = os.path.join(pdb_dir, pdb_file[0])

  curl_command = curl_template.format(dynamut2_link=dynamut2_link, pdb_file=pdb_file, mut_list=mut_list)
  !{curl_command} > 'curl_output.txt'
  with open('curl_output.txt') as f:
    res = f.read()
  if "200 OK" not in res:
    print(f'Nepavyko paleisti DynaMut2 {seq_name} sekai!')
    print(f'Atsakymas iš curl:\n{res}')
    continue
  if '"error":' in res:
    print('DynaMut2 grąžino klaidą {seq_name} sekai:')
    print(res)
    continue
  job_id = res.split('"job_id": "')[1].split('"}')[0]
  curl_job_ids[seq_name] = job_id

  sleep(0.5)

```

```python
curl_job_ids_file = 'dynamut2_job_ids.txt'
with open(curl_job_ids_file, 'w') as f:
  f.write('\n'.join([f'{k},{v}' for k, v in curl_job_ids.items()]))

print('\nUžklausos išsiųstos, liko laukti kol DynaMut2 pabaigs skaičiavimus')
print(f'Užklausų sąrašas: {curl_job_ids_file}')
```

# Palaukti ~10 min ir leisti cėlę apačioje

```python
dynamut2_link = "https://biosig.lab.uq.edu.au/dynamut2/api/prediction_mm"
curl_job_ids_file = 'dynamut2_job_ids.txt'
curl_get_template = "curl {dynamut2_link} -X GET -F job_id={job_id}"

with open(curl_job_ids_file, 'r') as f:
  curls_to_get = f.read().splitlines()
curls_to_get = {x.split(',')[0]: x.split(',')[1] for x in curls_to_get}

dynamut2_predictions = {}

for seq_name, job_id in curls_to_get.items():
  curl_command = curl_get_template.format(dynamut2_link=dynamut2_link, job_id=job_id)
  !{curl_command} > 'curl_get_output.json'
  with open('curl_get_output.json', 'r') as f:
    res = json.load(f)
  res_dict = res[list(res.keys())[0]]
  if 'prediction' in res_dict:
    dynamut2_predictions[seq_name] = res_dict['prediction']  # arba sum_ddg?
  else:
    print(f'Nėra rezultatų {seq_name}, gal reikia dar palaukti')
    continue

with open('dynamut2_results.csv', 'w') as f:
  f.write('\n'.join([f'{k},{v}' for k, v in dynamut2_predictions.items()]))
```
